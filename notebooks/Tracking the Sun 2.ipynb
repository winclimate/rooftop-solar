## Rebate or Grant Analysis

```python
file_path = '/Users/sienafrost/Downloads/public_datafile/TTS.csv'
df = pd.read_csv(file_path)
```

```python
df[df.zip_code == 501]
```

```python
#filter for just NY state
NY_state_df = df[df['state'] == 'NY']
```

```python
NY_state_df.head()
```

```python
unique_values = NY_state_df['rebate_or_grant'].value_counts().sort_index()

# Iterate through unique values and their counts
for value, count in unique_values.items():
    print(f"Value: {value}, Count: {count}")
```

```python
percentage_zero_values = (NY_state_df['rebate_or_grant'] == 0.0).mean() * 100
percentage_absent_values = (NY_state_df['rebate_or_grant'] == -1.0).mean() * 100

print(f"Percentage of responses that are 0.0: {percentage_zero_values:.2f}%")
print(f"Percentage of responses that are -1.0: {percentage_absent_values:.2f}%")
```

```python
condition = (NY_state_df['rebate_or_grant'] > 0.0) & (df['rebate_or_grant'] < 100.0)
percentage_values_between_0_and_100 = (condition).mean() * 100

print(f"Percentage of values between 0.0 and 100.0 in 'rebate_or_grant': {percentage_values_between_0_and_100:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(NY_state_df['rebate_or_grant'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Rebate or Grant Values')
plt.xlabel('Rebate or Grant Amount')
plt.ylabel('Frequency')
plt.show()
```

## Overall cost of projects

```python
unique_values_2 = NY_state_df['total_installed_price'].value_counts().sort_index()

# Iterate through unique values and their counts
for value, count in unique_values_2.items():
    print(f"Value: {value}, Count: {count}")
```

```python
percentage_zero_values_2 = (NY_state_df['total_installed_price'] == 0.0).mean() * 100
percentage_absent_values_2 = (NY_state_df['total_installed_price'] == -1.0).mean() * 100

print(f"Percentage of responses that are 0.0: {percentage_zero_values_2:.2f}%")
print(f"Percentage of responses that are -1.0: {percentage_absent_values_2:.2f}%")
```

```python
valid_rows_mask = (NY_state_df['total_installed_price'] != -1.0) & (NY_state_df['rebate_or_grant'] != -1.0)

# Calculate the 'price_after_rebate' values and set them using .loc
NY_state_df.loc[valid_rows_mask, 'price_after_rebate'] = (
    NY_state_df['total_installed_price'] - NY_state_df['rebate_or_grant']
)

# Display or use the DataFrame
NY_state_df_filtered = NY_state_df[valid_rows_mask]
NY_state_df_filtered.head()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(NY_state_df_filtered['price_after_rebate'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Price After Rebate')
plt.xlabel('Price After Rebate')
plt.ylabel('Frequency')
plt.show()
```

```python
mean_value = NY_state_df_filtered['price_after_rebate'].mean()
median_value = NY_state_df_filtered['price_after_rebate'].median()

# Display the mean and median
print(f"Mean of 'price_after_rebate': {mean_value:.2f}")
print(f"Median of 'price_after_rebate': {median_value:.2f}")

#mean>median
```

```python
cutoff = 70000.00
floor = -1000.00

# Create a histogram of the 'price_after_rebate' column with cutoff and floor
plt.figure(figsize=(8, 6))
plt.hist(
    NY_state_df_filtered['price_after_rebate'],
    bins=20,
    color='skyblue',
    edgecolor='black',
    range=(floor, cutoff)
)
plt.title('Distribution of Price After Rebate (with Cutoff and Floor)')
plt.xlabel('Price After Rebate')
plt.ylabel('Frequency')
plt.axvline(cutoff, color='red', linestyle='dashed', linewidth=2, label='Cutoff')
plt.axvline(floor, color='green', linestyle='dashed', linewidth=2, label='Floor')
plt.legend()
plt.show()

# Calculate the percentage of rows left off with the cutoff
percentage_left_off_cutoff = (NY_state_df_filtered['price_after_rebate'] > cutoff).mean() * 100

# Calculate the percentage of rows left off with the floor
percentage_left_off_floor = (NY_state_df_filtered['price_after_rebate'] < floor).mean() * 100

# Display the percentages
print(f"Percentage of rows left off with the cutoff: {percentage_left_off_cutoff:.2f}%")
print(f"Percentage of rows left off with the floor: {percentage_left_off_floor:.2f}%")
```

```python
valid_rows_mask_2 = (NY_state_df['total_installed_price'] != -1.0) & (NY_state_df['rebate_or_grant'] != -1.0)

# Calculate the percentage and create a new column 'percent_price_rebate' using .loc
NY_state_df.loc[valid_rows_mask_2, 'percent_price_rebate'] = (
    (NY_state_df['rebate_or_grant'] / NY_state_df['total_installed_price']) * 100
)

# Display or use the DataFrame with the new column
NY_state_df_filtered_2 = NY_state_df[valid_rows_mask_2]
NY_state_df_filtered_2.head()
```

```python
cutoff = 100.0

# Create a histogram of the 'percent_price_rebate' column with cutoff
plt.figure(figsize=(8, 6))
plt.hist(
    NY_state_df_filtered_2['percent_price_rebate'],
    bins=40,
    color='skyblue',
    edgecolor='black',
    range=(0, cutoff)
)
plt.title('Distribution of Percent Price Rebate (with Cutoff at 100%)')
plt.xlabel('Percent Price Rebate')
plt.ylabel('Frequency')
plt.axvline(cutoff, color='red', linestyle='dashed', linewidth=2, label='Cutoff')
plt.legend()
plt.show()

# Calculate the percentage of rows removed because of the cutoff
percentage_removed = (
    (NY_state_df_filtered_2['percent_price_rebate'] > cutoff).mean() * 100
)

# Display the percentage of rows removed
print(f"Percentage of rows removed because of the cutoff at {cutoff:.2f}%: {percentage_removed:.2f}%")
```

```python
import seaborn as sns
cutoff = 100.0

# Create a KDE plot with cutoff
plt.figure(figsize=(8, 6))
sns.kdeplot(NY_state_df_filtered_2['percent_price_rebate'], color='skyblue', fill=True)
plt.title('KDE Plot of Percent Price Rebate (with Cutoff at 100%)')
plt.xlabel('Percent Price Rebate')
plt.axvline(cutoff, color='red', linestyle='dashed', linewidth=2, label='Cutoff')
plt.xlim(-10, 120)
plt.legend()
plt.show()
```
