```python
import pandas as pd

# Load CSV files
sunroof_data = pd.read_csv('project-sunroof-postal_code.csv')
tts_data = pd.read_csv('TTS_LBNL_public_file_07-Sep-2022_all.csv')

# Create total_system_size_by_zip dictionary
total_system_size_by_zip = tts_data.groupby('zip_code')['system_size_DC'].sum().to_dict()

# Create main_table list of dicts
main_table = []
for _, row in sunroof_data.iterrows():
    zip_code = row['region_name']
    state_name = row['state_name']
    kw_total = row['kw_total'] if pd.notna(row['kw_total']) else 0
    total_system_size = total_system_size_by_zip.get(zip_code, 0)
    main_table.append({
        'state_name': state_name,
        'zip_code': zip_code,
        'kw_total': kw_total,
        'total_system_size': total_system_size,
        'gap': kw_total - total_system_size
    })

# Calculate average_gap_by_state_name dictionary
average_gap_by_state_name = {}
for item in main_table:
    state = item['state_name']
    average_gap_by_state_name.setdefault(state, []).append(item['gap'])
average_gap_by_state_name = {k: sum(v) / len(v) for k, v in average_gap_by_state_name.items()}

# Print the average gaps for New York and Massachusetts
print(f"Average Gap for New York: {average_gap_by_state_name.get('New York', 'N/A')}")
print(f"Average Gap for Massachusetts: {average_gap_by_state_name.get('Massachusetts', 'N/A')}")

# Print main_table in a table format
print(pd.DataFrame(main_table).to_string(index=False))
```
